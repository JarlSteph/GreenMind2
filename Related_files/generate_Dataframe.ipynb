```python
from transformers import GPT2Tokenizer

# Initialize tokenizer for token count
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Standardized column names
columns = [ "Prompt", "Number of tokens", "creative", "reasoning", "constraint", "contextual", "complexity"]

def count_tokens(text):
    return len(tokenizer.encode(text)) if isinstance(text, str) else 0

### Process Extended Coding Prompts (df)
df["Number of tokens"] = df["Prompt"].apply(count_tokens)
# df["creative"] = df.get("creative", 0)  # Default to 0 if not available
# df["reasoning"] = df.get("reasoning", 0)
# df["constraint"] = df.get("constraint", 0)
# df["contextual"] = df.get("contextual", 0)
difficulty_mapping = {
    "Easy": 1,
    "Intermediate": 2,
    "Hard": 3,
}
df["Difficulty"] = df["Difficulty"].replace(difficulty_mapping)

df.rename(columns={"Difficulty": "complexity"}, inplace=True)




```

```python
data = df3.copy()
# Function to extract user prompts
def extract_user_prompt(text_list):
    for item in text_list:
        if item['role'] == 'user' and item['content']:
            return item['content']
    return None  # Return None if no valid prompt is found

# Convert JSON-like structure into DataFrame
formatted_data = []
for entry in data:
    formatted_data.append({
        "id": entry["idx"],
        "Prompt": extract_user_prompt(entry["text"]),
        "Problem Type": entry["problem_type"],
        "Sub Type": entry["sub_type"],
        "Educational Stage": entry["educational_stage"],
        "Difficulty Level": entry["difficulty"]["level"],
        "Complexity": entry["difficulty"]["complexity"],
        "Explanation": entry["difficulty"]["explanation"],
        "Techniques": ", ".join(entry["techniques"]),
        "Knowledge Points": ", ".join(entry["knowledge_points"]),
        "Learning Objectives": ", ".join(entry["learning_objectives"]),
        "Solution Approach": entry["solution_approach"]["method"],
        "Solution Steps": " | ".join(entry["solution_approach"]["steps"]),
        "Common Pitfalls": " | ".join(entry["solution_approach"]["common_pitfalls"]),
        "Verification Method": entry["verification_method"],
        "Recommended Use": entry["recommended_use"]
    })

# Create DataFrame
df_final = pd.DataFrame(formatted_data)
```

```python
df_final[['Prompt', 'Complexity']]
```

```python
def normalize_difficulty(value):
    if value <= 3:
        return 1  # Easy
    elif 4 <= value <= 6:
        return 2  # Medium
    else:
        return 3  
df_final = df_final[['Prompt', 'Complexity']]
df_final.rename(columns={"Complexity": "complexity"}, inplace=True)
df_final["Number of tokens"] = df_final["Prompt"].apply(count_tokens)
df_final["complexity"] = df_final['complexity'].apply(normalize_difficulty) 
```
